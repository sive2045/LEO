```python
import pandas as pd

data = pd.read_csv('ConstellationEphemerides.csv')
```

```python
data.columns
```

```python
starlink_data = data[ ['Semimajor axis', 'Eccentricity', 'Inclination', 'Right Ascension of Ascending Node', 'Argument of Perigee', 'True Anomoly']].to_numpy()
```

```python
def OrbitCreator(dataslice):
    semimajor = dataslice[0] << u.AU
    ecc = dataslice[1] << u.one
    inc = dataslice[2] << u.deg
    raan = dataslice[3] << u.deg
    argp = dataslice[4] << u.deg
    nu = dataslice[5] << u.deg
    orb = Orbit.from_classical(Earth, semimajor, ecc, inc, raan, argp, nu)
    return orb
```

```python
orbits = [OrbitCreator(starlink_data[i]) for i in range(len(starlink_data))]
```

```python
frame = OrbitPlotter3D()
frame.plot_body_orbit(Earth, orbits[0])
```
